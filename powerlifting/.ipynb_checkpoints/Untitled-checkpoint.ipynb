```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
powerlifting = pd.read_csv("openpowerlifting.csv")
```

```python
print "Names:"
print powerlifting.columns
```

```python
print "Dimensions:"
print powerlifting.shape
```

```python
print "Number of unique competitors: %d" % (len(pd.unique(powerlifting.Name)))
```

```python
# Get some stats on the age
print "Unique Ages:"
print np.unique(powerlifting.Age)
ages = powerlifting.Age.dropna()
```

```python
print "Age of youngest competitor: %s" % (np.min(ages))
print "Age of oldest competitor: %s" % (np.max(ages))
```

```python
sns.set_style('white')

age_hist = sns.distplot(ages)
age_hist.set_title("Kernel Density Estimate (KDE) of Age")
age_hist.set_ylabel("")
plt.show()
```

```python
gender_plot = sns.countplot(powerlifting.Sex)
gender_plot.set_title("Powerlifting Gender Count")
gender_plot.set_ylabel("")
plt.show()
```
